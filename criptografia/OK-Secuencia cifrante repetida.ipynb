## Desafío
En este desafío, el servidor proporciona 5 textos cifrados con un algoritmo de cifrado de flujo (no importa cual). De ellos, tres están cifrados con la misma secuencia cifrante.

El desafío es descifrarlos y encontrar la secuencia cifrante.

Los textos claros cifrados con la secuencia cifrante repetida tienen las siguientes características.

- Están compuestos por caracteres ASCII imprimibles, sin espacios en blanco ni fines de línea. Esto implica que sus códigos están entre 33 y 126 inclusive.
Dos textos están compuestos por repeticiones de un mismo carácter. Por ejemplo, “AAAAAAAAAAAAAA” o “))))))))))))))))))))”
- Todos los textos tienen la misma longitud.
- Uno de los textos es la representación decimal de un número par. Por ejemplo “12345678”

Para obtener los textos cifrados, se debe hacer un requerimiento GET a una URL de la forma:

https://ciberseguridad.diplomatura.unc.edu.ar/cripto/stream/<email>/challenge/<id>
        
dónde <email> debe ser reemplazado por una dirección de correo electrónico registrada, y <id> por un número de mensaje válido (0,1,2,3 ó 4).

La respuesta será una secuencia de bytes codificada en base64.

La URL

https://ciberseguridad.diplomatura.unc.edu.ar/cripto/stream/<email>/challenge
devuelve los mismos textos cifrados, pero en un solo requerimiento. La respuesta contiene los cinco textos cifrados, uno por línea, y codificados individualmente en base64.

```python
def validate_same_letter(lista):
  return all(i == lista[0] for i in lista)
```

```python
def xor_between_list(lista_a, lista_b):
  if len(lista_a) != len(lista_b):
    raise ValueError("listas tiene que ser de mismo tamanio")
  xor_list = bytearray()
  for i in range(0, len(lista_a)):
    xor_list.append(lista_a[i] ^ lista_b[i])
  return xor_list
```

```python
def validate_numero_y_par(lista):
  for each in lista:
    if each < 48 or each > 57:
      return False
  if lista[-1] in [50,52,54,56]:
    return True
  return False
```

```python
def validate_output(keys):
  lista_encriptados = list(texto_encriptado_bytes_list)
  for each in lista_encriptados:
    texto_claro = xor_between_list(list(each), list(keys))

    test = ''
    for a in texto_claro:
      test = test + chr(a)
    print(test)
```

```python
def is_validate_ascii_imprimible(lista):
  for each in lista:
    if each < 33 or each > 126:
      return False
  return True
```
