```python
def xor_consecutivos(lista):
  new_string = []
  for each in range(0, len(lista), 4):
    cn = lista[each:each+2]
    cn1 = lista[each+2:each+4]
    cn_xor = int(cn,16) ^ int(cn1,16)
    new_string.append(hex(cn_xor))
  return new_string
```

```python
cipher = xor_consecutivos(message_decoded_in_hex)
print(str(cipher))
```

```python
def xor_two_list(lista_a, lista_b):
  if len(lista_a) != len(lista_b):
    raise ValueError("listas tiene que ser de mismo tamanio")
  new_string = []
  for index in range(0, len(lista_a)):
    cn = lista_a[index]
    cn1 = lista_b[index]
    cn_xor = int(cn,16) ^ int(cn1,16)
    new_string.append(hex(cn_xor))
  return new_string
```

```python
declared_text_in_hex_list = [declared_text_in_hex[i:i+2] for i in range(0, len(declared_text_in_hex), 2)]
print(str(declared_text_in_hex_list))
```

t1 ^ c1^c2 = k12

t2 ^ c3^c4 = k34


```python
k_posible = []
for index in range(0, len(cipher)-len(declared_text_in_hex_list)):
  ci = cipher[index:index+17]
  k_ = xor_two_list(ci , declared_text_in_hex_list)
  if k_[0] == k_[16]:
    k = k_[:len(k_)-1]
    k_index = index % 16
    k_reorder = k[-k_index:] + k[:-k_index]
    k_posible.append(k_reorder)
    print(str(k_reorder))
```

t1 = k12 ^ c1^c2

t2 = k34 ^ c3^c4

```python
def get_texto(cipher, k):
  str = ''
  for index in range(0,len(cipher), 16):
    des = cipher[index:index+16]
    t = xor_two_list(des, k[0:len(des)])
    for each in t:
      str= str +chr(int(each, 16))
  return str
```

```python
f = open("message.txt", "w")
for each_k in k_posible:
  texto_claro_total = get_texto(cipher,each_k)
  if texto_claro_total.find("user@example.com>") > 0:
    print("k: "+str(each_k))
    print("texto claro: \n"+texto_claro_total)
    f.write(texto_claro_total)
    break
f.close()
```
